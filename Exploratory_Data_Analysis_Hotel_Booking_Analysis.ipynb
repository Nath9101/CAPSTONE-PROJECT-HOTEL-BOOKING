Cleaning the data

```python
# Creating a copy of original dataset
hotel_df=df.copy()
```

Columns like country,agent,children and company have high percentage of null values. Above all of these columns agent and company have high percentage of null values.

```python
# Deleting agent and company columns
df=df.drop(['agent','company'],axis=1)
```

```python
# Dropping all null values of columns children and country
df=df.dropna(axis=0)
```

```python
# Checking null values if there are any
df.isnull().sum()
```

```python
# Describing the dataframe
df.describe()
```

```python
df1=df.copy()
```

```python
df1['children'].value_counts()
```

```python
# Removing strange values from columns
df1.loc[df1['lead_time']>500,'lead_time']=500
df1.loc[df1['adults']>4,'adults']=4
```

```python
# Creating a new columns 'total_members' by adding 'children' and 'adults'
df1['total_members']=df1['children']+df1['adults']
```

```python
# Creating a new columns 'total_stay' by adding 'stays_in_weekend_nights' and 'stays_in_week_nights'
df1['total_stay']=df1['stays_in_weekend_nights']+df1['stays_in_week_nights']
```

Visualising Data

```python
hotel_df.describe()
```

```python
# Correlation between all variables
plt.figure(figsize=(20,10))
sns.heatmap(df1.corr(),annot=True)
```

1. Best time to book a room in hotels

```python
plt.figure(figsize=(16,8))
sns.lineplot(x='arrival_date_month',y='adr',hue='hotel',data=df1)
```

1. In city hotels, the average rate is more expensive during may, june, july and august month.
2. In resort, the average rate is more expensive during july and august.

2. What is the best meal packages available in hotels?

```python
best_meal=df1['meal'].value_counts()
best_meal.plot(kind='bar')
plt.title('Meal packages')
plt.ylabel('count')
plt.xlabel('Meal type')
print(f'BB-Bed and Breakfast\n HB-Halfboard\n SC-Self Catering\n FB-Full Board')
```

1. Majority of guest consume bed and breakfast.

3. Cancellation rate in city hotels and resort hotels.

```python
plt.figure(figsize=(16,6))
sns.countplot(x='hotel',hue='is_canceled',data=df1)
plt.title("cancellation rates in city hotels and resotrt hotels")
plt.show()
```

1. Percentage of cancellation in resort hotel is less compared to city hotel.

4. Number of customer repeating their hotel bookings.

```python
plt.figure(figsize=(16,10))
sns.countplot(data=hotel_df,x='is_repeated_guest').set_title("custome repeating their bookings",fontsize=18)
```

1. Number of repeated guests are very low.

5. Month which is mostly occupied vs month which is least occupied



```python
month_df=hotel_df[hotel_df['is_canceled']==0]['arrival_date_month'].value_counts().reset_index().rename(columns={'index':'month','arrival_date_month':'number_of_bookings'})
```

```python
plt.figure(figsize=(16,10))
ax=sns.barplot(x='month',y='number_of_bookings',data=month_df)
```

1. August month is mostly occupied by the guest.
2. January month is least occupied by the guest.

6. Percentage of bookings in city hotel and resort.



```python
labels=hotel_df['hotel'].value_counts().index.tolist()
sizes=hotel_df['hotel'].value_counts().tolist()
explode=(0,0.05)
colors=['red','blue']
```

```python
#pie plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',startangle=90,textprops={'fontsize':14})
```

1. City hotels are mostly booked by the guest.
2. resort hotels are least booked by the guest.
